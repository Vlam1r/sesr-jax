```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['text.usetex'] = True
plt.rcParams['font.size'] = 12
plt.rcParams['axes.linewidth'] = 1.5
```

```python
psnr_vals = pd.read_csv("pruning_params/psnr.csv")
psnr_vals = np.array(psnr_vals["robust-puddle-9 - psnr"])
psnr_vals = psnr_vals.reshape(12, 50)
```

```python
sparsities = [round(100*(1 - 0.8 ** i), 1) for i in range(11)]
```

```python
sparsities
```

```python
fig, (ax2, ax1) = plt.subplots(1, 2, figsize=(12, 5))
ax1.plot(psnr_vals[0], label="0.0\% Sparse")
ax1.plot(psnr_vals[1], label="20.0\% Sparse")
ax1.plot(psnr_vals[3], label="48.8\% Sparse")
ax1.plot(psnr_vals[6], label="73.8\% Sparse")
ax1.plot(psnr_vals[10], label="89.3\% Sparse")
ax1.legend()
ax1.set_ylabel("Test Set PSNR")
ax1.set_xlabel("Epoch")
ax1.set_ylim(33.8,35)

ax2.plot(psnr_vals[0], label="0.0\% Sparse")
ax2.plot(psnr_vals[1], label="20.0\% Sparse")
ax2.plot(psnr_vals[3], label="48.8\% Sparse")
ax2.plot(psnr_vals[6], label="73.8\% Sparse")
ax2.plot(psnr_vals[10], label="89.3\% Sparse")
ax2.legend()
ax2.set_ylabel("Test Set PSNR")
ax2.set_xlabel("Epoch")
ax2.set_ylim(32,35)
plt.savefig("Pruning Training Curves", dpi=200)
```
